```python
from sklearn.model_selection import StratifiedShuffleSplit
split=StratifiedShuffleSplit(n_splits=1,test_size=0.3,random_state=42)
for x_train,x_test in split.split(df,df['chas']):
    x_train=df.loc[x_train]
    x_test=df.loc[x_test]
   
```

```python
from pandas.plotting import scatter_matrix
attri=['medv','rm','zn','lstat']
scatter_matrix(df[attri],figsize=(15,10))
```

```python
x_train1=x_train.drop('medv',axis=1)
y_train1=x_train.medv
x_test1=x_test.drop('medv',axis=1)
y_test1=x_test.medv
from sklearn.preprocessing import StandardScaler
stand=StandardScaler()
x_train1=stand.fit_transform(x_train1)
x_test1=stand.fit_transform(x_test1)
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train1,y_train1)
```

```python
yp=list(model.predict(x_test1))
```

```python
import numpy as np
from sklearn.metrics import mean_squared_error
m=mean_squared_error(y_test1,yp)
np.sqrt(m)
```
